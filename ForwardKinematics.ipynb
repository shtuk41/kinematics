```python
import numpy as np
Q1 = 45
Q1R = np.deg2rad(Q1)
c1 = np.cos(Q1R)
s1 = np.sin(Q1R)
a1 = 10

Q2 = 20
Q2R = np.deg2rad(Q2)
c2 = np.cos(Q2R)
c2 = np.cos(Q2R)
s2 = np.sin(Q2R)
a2 = 10



A1 =np.array(   [[c1, -s1, 0, a1*c1],
                [ s1,  c1, 0, a1*s1],
                [ 0,   0,  1, 0],
                [ 0,   0,  0, 1]])

A2 =np.array([  [c2, -s2,  0, a2*c2],
                [ s2,  c2, 0, a2*s2],
                [ 0,   0,  1, 0],
                [ 0,   0,  0, 1]])

L = np.array([1,0,0,1])

L2 = A1 @ A2 @ L

```

```python
import copy
from mpl_toolkits import mplot3d

%matplotlib ipympl

import numpy as np
import matplotlib.pyplot as plt

def plot_star(ax, center, color='blue', size=100, label=None):
    ax.scatter(center[0], center[1], center[2], color=color, s=size, marker='o')
    if label:
        ax.text(center[0] + 0.5, center[1], center[2], label, color='black', fontsize=12)
    

fig = plt.figure()
plt.axis('equal')
ax = plt.axes(projection='3d')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Simple rotation')

link1 = A1 @ np.array([0,0,0,1])

xline1 = [0,link1[0]]
yline1 = [0,link1[1]]
zline1 = [0,link1[2]]

ax.plot3D(xline1, yline1, zline1, 'green')

link2 = A1 @ A2 @ np.array([0,0,0,1])

xline2 = [link1[0],link2[0]]
yline2 = [link1[1],link2[1]]
zline2 = [link1[2],link2[2]]

ax.plot3D(xline2, yline2, zline2, 'red')

plot_star(ax, [0,0,0], color = 'blue', label = 'JF1')
plot_star(ax, link1, color = 'blue', label = 'JF2')
plot_star(ax, link2, color = 'blue', label = 'Tool')

ax.view_init(elev=90, azim=90)
plt.show()
```
